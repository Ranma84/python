**Los operadores de asignación aumentada**


| Operación | Equivalencia |
|-----------|--------------|
| a = a + 1 | a += 1       |
| a = a - 1 | a -= 1       |
| a = a * 1 | a *= 1       |
| a = a / 1 | a /= 1       |
| a = a % 1 | a %= 1       |

# Otra de las ventajas que Python nos provee, es la de poder asignar en una sola instrucción, múltiples variables:
En una sola instrucción, estamos declarando tres variables: a, b y c y asignándoles un valor concreto a
cada una

```python
a = 4
print(a)
a = a + 5
print(a)
a += 5
print(a)
```
